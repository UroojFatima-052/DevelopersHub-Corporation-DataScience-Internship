```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('insurance.csv') # laoding the dataset 
```

## Visualization

#### Age Distribution

```python
sns.histplot(data=df, x='age', bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')

plt.show()
```

#### What's the impact of Age on Claim Amount ?

```python
sns.scatterplot(data=df, x='age', y='charges')
plt.title('Age vs. Charges (Claim Amount)')
plt.xlabel('Age')
plt.ylabel('Charges')

plt.show()
```

#### What's the impace of BMI on Claim Amount ?

```python
sns.scatterplot(data=df, x='bmi', y='charges')
plt.title('BMI vs. Charges (Claim Amount)')
plt.xlabel('BMI')
plt.ylabel('Charges')

plt.show()
```

#### What's the impact of smoking on claim amount ?

```python
sns.boxplot(data=df, x='smoker', y='charges')
plt.title('Smokers vs. Claim Amount')
plt.xlabel('Smoker')
plt.ylabel('Charges')
plt.show()
```

#### Does no.of childrens affect Claim Amount ?

```python
sns.scatterplot(data=df, x='children', y='charges')
plt.title('No.of Children vs. Claim Amount')
plt.xlabel('No.of Children')
plt.ylabel('Claim Amount')
plt.show()
```

#### Heatmap

```python
sns.heatmap(df.select_dtypes('number').corr(), annot=True, cmap='Blues')
plt.show()
```
