## Basic Data Exploration

In this step I have explore key customer characteristics such as age, job type, and marital status to understand how they relate to personal loan acceptance. This helps identify which customer groups are more likely to accept a loan offer.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('bank-additional-full.csv', sep=';') # loading the dataset using ';' delimeter
```

## Age Distribution

```python
sns.histplot(data=df, x='age')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')

plt.show()
```

## Age vs. Loan Acceptance

```python
sns.boxplot(data=df, x='y', y='age')
plt.title('Age vs. Loan Acceptance')
plt.xlabel('Loan Acceptance')
plt.ylabel('Age')

plt.show()
```

## Job Distribution

```python
plt.figure(figsize=(15, 5))

sns.countplot(data=df, x='job', palette='coolwarm')
plt.title('Job Distribution')
plt.xlabel('Job')
plt.ylabel('Count')

plt.show()
```

## Job vs. Loan Acceptance

```python
sns.countplot(data=df, y='job', hue='y')
plt.title('Job vs. Loan Acceptance')
plt.xlabel('Count')
plt.ylabel('Job Type')

plt.show()
```

## Marital Status Distribution

```python
sns.countplot(data=df, x='marital', palette='Set2')
plt.title('Marital Status Distribution')
plt.xlabel('Marital Status')
plt.ylabel('Count')

plt.show()
```

## Marital Status vs. Loan Acceptance

```python
sns.countplot(data=df, x='marital', hue='y')
plt.title('Marital Status vs. Loan Acceptance')
plt.xlabel('Marital Status')
plt.ylabel('Count')

plt.show()
```
