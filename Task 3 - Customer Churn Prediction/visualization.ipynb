```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cleaned_churn_dataset.csv') # loaded the cleaned dataset
```

# Visualization

## Churn Distribution

```python
sns.countplot(data=df, x='Churn', hue='Churn', palette='Greens' )
plt.title('Churn Distribution')
plt.xlabel('Churn (0 = Stayed, 1 = Churned)')
plt.show()
```

## Age Distribution

```python
sns.histplot(data=df, x='Age')

plt.title('Age Distribution')
plt.show()
```

## Age vs. Churn

```python
sns.boxplot(data=df, x='Churn', y='Age')

plt.title('Which age of people most likely to Churn ? ')
plt.xlabel('Churn (0 = Stayed, 1 = Left)')
plt.ylabel('Age')
plt.show()
```

## Tenure vs. Churn

```python
sns.boxplot(data=df, x='Churn', y='Tenure')

plt.title('Does Tenure affect Churn ? ')
plt.xlabel('Churn (0 = Stayed, 1 = Left)')
plt.ylabel('Tenure (in months)')
plt.show()
```

## Number of loan vs. Churn

```python
sns.boxplot(data=df, x='Churn', y='No_Loan')

plt.title('Does loan affects the Churn ?')
plt.xlabel('Churn (0 = Stayed, 1 = Left)')
plt.show()
```

## Total Transactions vs. Churn

```python
sns.boxplot(data=df, x='Churn', y='Total_trans_no')
plt.title('Does Total Transactions affects Churn ?')
plt.xlabel('Churn (0 = Stayed, 1 = Left)')
plt.ylabel('Total Transactions')
plt.show()
```
