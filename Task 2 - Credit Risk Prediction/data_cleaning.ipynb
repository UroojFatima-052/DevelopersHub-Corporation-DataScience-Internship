### CLEANING

#### Step : 01
I have drop Upfront_charges, Interest_rate_spread, and rate_of_interest columns because they have a lot of missing values and are not core features for predicting default.

```python
df.drop(columns=['Upfront_charges', 'rate_of_interest', 'Interest_rate_spread'], inplace=True)
```

#### Step : 02

For numeric columns we fill missing values with the median, and for categorical columns we fill with the most frequent value (mode) so there are no NaNs left in the dataset.

```python
# Handling Numerical Values

numerical_columns = ['property_value', 'LTV', 'dtir1', 'income']

for col in numerical_columns:
    df[col] = df[col].fillna(df[col].median())
```

```python
# Handling Categorical Columns

categorical_columns = ['loan_limit', 'approv_in_adv', 'submission_of_application', 'age', 'loan_purpose', 'Neg_ammortization', 'term']

for col in categorical_columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

#### Step : 03

saving the cleaned dataset into a new file.

```python
df.to_csv('cleaned_loan_default.csv', index=False)
```
